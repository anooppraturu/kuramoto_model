```python
import numpy as np
import matplotlib.pyplot as plt
import sync as sync
```

```python
#initialize network with uniform coupling strength
network = sync.osc_network(200, 2.5)
```

```python
#integrate trajectories and compute time averaged order parameters
network.time_avg_order_params()
```

```python
#plot synchronization over time
fig, ax = plt.subplots(figsize=(10,5))

for i in np.arange(200):
    ax.plot(np.arange(4000), network.trajectory[:,i], c='black', alpha=0.5, lw=0.1)
    
ax.set_ylabel('$\phi_i$', fontsize=15)
ax.set_xlabel('Time', fontsize=15)
```
